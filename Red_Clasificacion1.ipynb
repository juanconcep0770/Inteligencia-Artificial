```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

# Generar los datos de entrenamiento
np.random.seed(0)
num_samples = 1000
color1 = np.random.normal(loc=0, scale=1, size=(num_samples, 2))  # Color 1 (por ejemplo, azul)
color2 = np.random.normal(loc=3, scale=1, size=(num_samples, 2))  # Color 2 (por ejemplo, rojo)

x_train = np.vstack((color1, color2))
y_train = np.vstack((np.ones((num_samples, 1)), np.zeros((num_samples, 1))))

# Crear el modelo de la red neuronal
model = tf.keras.Sequential([
    tf.keras.layers.Dense(16, activation='relu', input_shape=(2,)),
    tf.keras.layers.Dense(16, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

# Compilar el modelo
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Entrenar el modelo
model.fit(x_train, y_train, epochs=50, batch_size=32)

# Generar datos de prueba
color1_test = np.random.normal(loc=0, scale=1, size=(num_samples, 2))  # Color 1 (por ejemplo, azul)
color2_test = np.random.normal(loc=3, scale=1, size=(num_samples, 2))  # Color 2 (por ejemplo, rojo)

x_test = np.vstack((color1_test, color2_test))
y_test = np.vstack((np.ones((num_samples, 1)), np.zeros((num_samples, 1))))

# Evaluar el modelo en los datos de prueba
loss, accuracy = model.evaluate(x_test, y_test)
print("Loss:", loss)
print("Accuracy:", accuracy)

# Realizar predicciones con el modelo entrenado
predictions = model.predict(x_test)

# Graficar los resultados
plt.scatter(x_test[:, 0], x_test[:, 1], c=predictions.ravel(), cmap='coolwarm')
plt.colorbar()

plt.title("Después de la red neuronal")
plt.show()

```

```python
import numpy as np
from minisom import MiniSom

# Preparar los datos (ejemplo)
data = np.random.rand(500, 14)  # Datos de ejemplo: 100 instancias con 10 características
```

```python
# Configurar el mapa SOM
grid_size = (15, 15)  # Tamaño de la cuadrícula del mapa SOM
input_len = data.shape[1]  # Número de características en tus datos

som = MiniSom(grid_size[0], grid_size[1], input_len)

# Entrenar el mapa SOM
som.train_random(data, num_iteration=200)  # Entrena el mapa SOM con 100 iteraciones
```

```python
import matplotlib.pyplot as plt

# Visualizar el mapa SOM
plt.figure(figsize=(10, 10))
plt.pcolor(som.distance_map().T, cmap='bone_r')  # Mapa de calor de las distancias entre las neuronas
plt.colorbar()

# Agregar los datos al mapa SOM
for i, x in enumerate(data):
    w = som.winner(x)  # Neurona ganadora para cada instancia de datos
    plt.plot(w[0] + 0.5, w[1] + 0.5, 'o', markerfacecolor='None', markeredgecolor='r', markersize=10, markeredgewidth=2)

plt.show()
```

```python
import matplotlib.pyplot as plt

# Obtén las coordenadas de las neuronas del mapa SOM
neuron_positions = som.get_weights()

# Extrae las coordenadas X e Y de las neuronas
x_values = [pos[0] for pos in neuron_positions]
y_values = [pos[1] for pos in neuron_positions]

# Crea el diagrama de dispersión
plt.scatter(x_values, y_values)
plt.xlabel('Coordenada X')
plt.ylabel('Coordenada Y')
plt.title('Mapa SOM en Diagrama de Dispersión')
plt.show()
```

```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt

# Generar los datos de entrenamiento
x_train = np.linspace(0, 2*np.pi, 1000)  # Valores de entrada
y_train = np.sin(x_train)                # Valores objetivo (seno de los valores de entrada)

# Crear el modelo de la red neuronal
model = tf.keras.Sequential([
    tf.keras.layers.Dense(20, activation='relu', input_shape=(1,)), #defino una capa densa (fully connected) de una red neuronal en TensorFlow.
    tf.keras.layers.Dense(20, activation='relu'),                   #input_shape=(1,): Especifica la forma de los datos de entrada a la capa. 
    tf.keras.layers.Dense(1)                                        # En este caso, la capa espera datos de entrada unidimensionales con una única característica.
])

# Compilar el modelo
model.compile(optimizer='adam', loss='mse')

# Entrenar el modelo
model.fit(x_train, y_train, epochs=150, batch_size=32)

# Generar los datos de prueba
x_test = np.linspace(0, 2*np.pi, 100)
y_test = np.sin(x_test)

# Realizar predicciones con el modelo entrenado
y_pred = model.predict(x_test)

# Graficar los resultados
plt.plot(x_test, y_test, label='Función objetivo')
plt.plot(x_test, y_pred, label='Predicción de la red neuronal')
plt.legend()
plt.show()
```
